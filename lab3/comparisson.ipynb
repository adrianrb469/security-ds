```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import roc_curve, auc

# Cargar los resultados de ambos modelos
# Modelo 1: Random Forest con TF-IDF
model1_metrics = {
    'Accuracy': 0.9468,
    'Precision': 0.9886,
    'Recall': 0.9039,
    'F1': 0.9444,
    'ROC AUC': 0.9829,
    # second metrics
    'CV Accuracy': 0.9355,
    'CV Precision': 0.9839,
    'CV Recall': 0.8856,
    'CV F1': 0.9317,
    'CV ROC AUC': 0.9849
}

# Modelo 2: Red Neuronal con Embeddings
model2_metrics = {
    'Accuracy': 0.9611,
    'Precision': 0.9659,
    'Recall': 0.9558,
    'F1': 0.9608,
    'ROC AUC': 0.9837,
    # second metrics
    'CV Accuracy': 0.9572,
    'CV Precision': 0.9687,
    'CV Recall': 0.9456,
    'CV F1': 0.9566,
    'CV ROC AUC': 0.9885
}

# Crear tabla comparativa
metrics = ['Accuracy', 'Precision', 'Recall', 'F1', 'ROC AUC',
           'CV Accuracy', 'CV Precision', 'CV Recall', 'CV F1', 'CV ROC AUC']

comparison_data = {
    'Métrica': metrics,
    'Modelo 1 (Random Forest + TF-IDF)': [model1_metrics[m] for m in metrics],
    'Modelo 2 (Red Neuronal + Embeddings)': [model2_metrics[m] for m in metrics]
}

comparison_df = pd.DataFrame(comparison_data)
print("Comparación de modelos:")
print(comparison_df)

# Guardar la comparación en un archivo CSV
comparison_df.to_csv('./model_results/model_comparison.csv', index=False)

# Visualizar la comparación
plt.figure(figsize=(12, 8))
comparison_df_plot = comparison_df.set_index('Métrica')
comparison_df_plot.plot(kind='bar', rot=45)
plt.title('Comparación de Métricas entre Modelos')
plt.ylabel('Valor')
plt.xlabel('Métrica')
plt.legend(title='Modelo')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('./model_results/model_comparison.png')

# Visualizar las curvas ROC de ambos modelos en una sola gráfica
plt.figure(figsize=(10, 8))

# Cargar las curvas ROC de ambos modelos
# En un entorno real, esto se cargaría de los
```
